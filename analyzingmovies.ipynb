```python
import numpy as np
def cal_desc_stats(series):
    mean = series.mean()
    median = np.median(series)
    stdev = np.std(series)
    return (mean, median,stdev)

(mc_mean,mc_median,mc_stdev) = cal_desc_stats(movies['Metacritic_norm_round'])
mc_mean,mc_median,mc_stdev
```

```python
(fg_mean,fg_median,fg_stdev) = cal_desc_stats(movies['Fandango_Stars'])
fg_mean,fg_median,fg_stdev
```

### correlation coeff between metacritic and fandango is low

```python
from scipy.stats import linregress
l = linregress(movies['Metacritic_norm_round'],movies['Fandango_Stars'])
l
```

```python
predicted_fandango_rating = 3 * l[0]  + l[1]
```

```python
predicted_fandango_rating
```

```python
meta_rating = np.arange(1,5.5,.5)
print(meta_rating,len(meta_rating))
```

```python
def predict_fandango_rating(mt):
    return mt * l[0] + l[1]
predicted_fandango_rating = [predict_fandango_rating(x) for x in meta_rating]
predicted_fandango_rating
```

```python
plt.scatter(movies['Metacritic_norm_round'],movies['Fandango_Stars'])
plt.plot(meta_rating,predicted_fandango_rating)
plt.xlim(1,5)
```
